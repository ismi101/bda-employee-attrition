## **5. Methodology** ##
- **5.1 Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- **5.2 Loading the Data**

```python
data = pd.read_csv("HR Employee Attrition.csv")
data.info()
```

```python
data.head()
```

- **5.3 Data Preprocessing**

```python
# Handling missing values: Drop rows with null values
data = data.dropna()

# Handling duplicates: Drop duplicate rows
data = data.drop_duplicates()

# Display information about the cleaned dataset
print("\nCleaned Dataset Info:")
print(data.info())
```

- **5.4 Results Analysis & Visualization**

**What is the average age of employees in the dataset, and what is the distribution of age across different job roles?**

Visualizations

```python
# Average Age of Employees
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age in the Dataset')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Age Distribution Across Different Job Roles
plt.figure(figsize=(14, 8))
sns.boxplot(x='JobRole', y='Age', data=data, palette='viridis', hue='JobRole', dodge=False)
plt.title('Age Distribution Across Different Job Roles')
plt.xlabel('Job Role')
plt.ylabel('Age')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.legend().set_visible(False)  # Hide the legend for better clarity
plt.show()
```

Analysis:

The average age of employees in the dataset appears to be centered around the late 20s to 40s age group.
The positive/right-skewed distribution suggests that a significant portion of the workforce falls within this age range.

The age distribution among job roles varies, with most job roles showing a similar age distribution.
Notable differences exist in job roles such as "Managers" and "Research Directors," where employees are primarily in the higher age range of 40s to 50s.
In contrast, the job role of "Sales Representatives" has a younger workforce, predominantly in the early 20s to 30s.

**Is there a relationship between the distance from home and the frequency of overtime work among employees, considering their work-life balance levels?**

Visualization

```python
# Custom color palette for the scatter plot
scatter_palette = {'No': 'blue', 'Yes': 'red'}

# Visualization: Relationship Between Distance from Home, Overtime, and Work-Life Balance
plt.figure(figsize=(12, 8))
sns.scatterplot(x='DistanceFromHome', y='WorkLifeBalance', hue='OverTime', data=data, palette=scatter_palette, alpha=0.7, edgecolor='w')
plt.title('Relationship Between Distance from Home, Overtime, and Work-Life Balance')
plt.xlabel('Distance from Home')
plt.ylabel('Work-Life Balance')

# Create a custom legend outside the plot with only two colors
legend_labels = list(scatter_palette.keys())
legend_colors = list(scatter_palette.values())

# Create a legend with correct colors
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', label=label,
                             markerfacecolor=color, markersize=10) for label, color in zip(legend_labels, legend_colors)]

plt.legend(handles=legend_handles, title='Overtime', loc='upper right')

plt.show()
```

```python
# Create distance bins
distance_bins = [0, 5, 10, 15, 20, 25, 30]
distance_labels = ['0-5', '6-10', '11-15', '16-20', '21-25', '26-30']

# Bin the 'DistanceFromHome' variable
data['DistanceCategory'] = pd.cut(data['DistanceFromHome'], bins=distance_bins, labels=distance_labels, include_lowest=True)

# Visualization: Bar graph for Work-Life Balance across Distance Categories
plt.figure(figsize=(12, 8))
sns.countplot(x='DistanceCategory', hue='WorkLifeBalance', data=data, palette='coolwarm')
plt.title('Work-Life Balance Across Distance Categories')
plt.xlabel('Distance Category')
plt.ylabel('Count')
plt.legend(title='Work-Life Balance', loc='upper right', labels=data['WorkLifeBalance'].unique().tolist())
plt.show()
```

Analysis

The analysis reveals a clear distribution pattern between the frequency of overtime work and work-life balance levels. Employees reporting the highest work-life balance tend to have the least amount of overtime, while the group with the lowest work-life balance has the highest incidence of overtime. A surprising finding is observed in the second-highest category of work-life balance, where the highest amount of employees report doing overtime. This unexpected pattern suggests that factors other than work-life balance may be influencing the decision to work overtime in this specific group.

Contrary to expectations, the distribution of work-life balance across different distances from home is relatively similar. This indicates that, based on the available data, distance from home may not be a significant factor influencing work-life balance.

**Given the current attrition rate and job satisfaction levels, what interventions or improvements in work-life balance and environment satisfaction could be recommended to reduce attrition among employees?**

Visualization

```python
# Visualization for Attrition 'Yes'
plt.figure(figsize=(14, 8))
sns.countplot(x='WorkLifeBalance', hue='EnvironmentSatisfaction', data=data[data['Attrition'] == 'Yes'],
              palette='viridis', hue_order=[1, 2, 3, 4], edgecolor='black', hue_norm=(0, 4))

plt.title('Work-Life Balance and Environment Satisfaction for Attrition (Yes)')
plt.xlabel('Work-Life Balance')
plt.ylabel('Count')
plt.legend(title='Environment Satisfaction', loc='upper right', labels=data['EnvironmentSatisfaction'].unique().tolist())
plt.show()
```

```python
# Visualization for Attrition 'No'
plt.figure(figsize=(14, 8))
sns.countplot(x='WorkLifeBalance', hue='EnvironmentSatisfaction', data=data[data['Attrition'] == 'No'],
              palette='viridis', hue_order=[1, 2, 3, 4], edgecolor='black', hue_norm=(0, 4))

plt.title('Work-Life Balance and Environment Satisfaction for Attrition (No)')
plt.xlabel('Work-Life Balance')
plt.ylabel('Count')
plt.legend(title='Environment Satisfaction', loc='upper right', labels=data['EnvironmentSatisfaction'].unique().tolist())
plt.show()
```

Analysis

Overall, both employees with attrition ("yes") and without attrition ("no") exhibit a similar pattern in work-life balance ratings. The majority of employees report a work-life balance of 3, followed by 2, 4, and 1. This suggests that work-life balance is a common concern for employees across attrition groups.

Employees without attrition ("no") display a clear pattern in environmental satisfaction, with most reporting either 1 or 4, followed by 2 or 3. In contrast, employees with attrition ("yes") show more variability, indicating that environmental satisfaction may be a more critical factor influencing attrition for this group.

Given the common pattern in work-life balance ratings for both attrition groups, organizations should prioritize interventions aimed at enhancing overall work-life balance. This may include flexible work arrangements, time management training, and policies promoting a healthy work-life equilibrium.

Analysis

A normal distribution pattern is observed for education levels and attrition. The highest attrition occurs at the middle education levels (3), while attrition is lower towards the extremes (1 and 5). This suggests that moderate education levels may be associated with higher attrition rates.

A clear pattern emerges with the number of companies worked and attrition. The attrition rate decreases as the number of companies worked increases. Employees who have worked in only one company exhibit the highest attrition rate, surpassing even those who have not worked in any company. This indicates a potential inverse relationship between the number of companies worked and attrition.

Potential Explanation: Employees with moderate education levels (e.g., Bachelor's degree) may be more susceptible to attrition due to factors such as career advancement opportunities, job market demand, or aspirations for change. Those at the extremes of education (Below College and Doctorate) may have clearer career paths or unique motivations that contribute to lower attrition rates.

**What is the distribution of job satisfaction levels and attrition rates among different education fields?**

Visualization

```python
# Stacked Bar Plot for EducationField, JobSatisfaction, and Attrition
plt.figure(figsize=(14, 8))
sns.countplot(x='EducationField', hue='JobSatisfaction', data=data, palette='viridis', edgecolor='black', hue_order=[1, 2, 3, 4])
plt.title('Distribution of Job Satisfaction Levels Across Education Fields')
plt.xlabel('Education Field')
plt.ylabel('Count')
plt.legend(title='Job Satisfaction', loc='upper right', labels=data['JobSatisfaction'].unique().tolist())
plt.show()
```

```python
# Stacked Bar Plot for EducationField and Attrition
plt.figure(figsize=(14, 8))
sns.countplot(x='EducationField', hue='Attrition', data=data, palette='pastel', edgecolor='black')
plt.title('Distribution of Attrition Rates Across Education Fields')
plt.xlabel('Education Field')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right', labels=data['Attrition'].unique().tolist())
plt.show()
```

Analysis

A consistent pattern is observed in job satisfaction levels across different education fields. The majority of employees report job satisfaction levels of 1 or 3, followed by 4 or 2. This suggests a tendency for employees to polarize towards extreme satisfaction (1 and 4) or moderate satisfaction (2 and 3). The "Life Sciences" and "Medical" fields emerge as the most popular choices among employees, while "Human Resources" and "Others" are less common.

Lower attrition rates are observed in the "Life Sciences," "Medical," and "Others" fields. In contrast, higher attrition rates are noted in the "Marketing," "Technical Degree," and "Human Resources" fields.

Potential Explanation: The consistent pattern in job satisfaction levels suggests that employees in certain education fields may have more uniform perceptions of job satisfaction. Higher job satisfaction, particularly in "Life Sciences," "Medical," and "Others," could contribute to lower attrition rates as employees are content with their roles.
